```python
def clean(data):
    data = data.drop(['LandSlope', 'TotRmsAbvGrd', 'BsmtQual', 'GrLivArea', 'LowQualFinSF', '1stFlrSF', '2ndFlrSF', 'FullBath', 'HalfBath', 'SaleType', 'SaleCondition', 'YrSold', 'MoSold', 'PoolQC', 'Fence', 'MiscFeature', 'FireplaceQu', 'Alley', 'BsmtHalfBath', 'BsmtFullBath', 'Id'],axis=1)
    data['LotFrontage'] =  data['LotFrontage'].fillna(data['LotFrontage'].median())
    data['MasVnrType']=data['MasVnrType'].fillna(data['MasVnrType'].value_counts().index[0])     
    data['MasVnrArea'] = data['MasVnrArea'].fillna(data['MasVnrArea'].median())
    data['BsmtExposure']=data['BsmtExposure'].fillna(data['BsmtExposure'].value_counts().index[0])
    data['BsmtCond']=data['MasVnrType'].fillna(data['MasVnrType'].value_counts().index[0])
    data['BsmtFinType1']=data['BsmtFinType1'].fillna(data['BsmtFinType1'].value_counts().index[0])
    data['BsmtFinSF1']=data['BsmtFinSF1'].fillna(data['BsmtFinSF1'].median())
    data['BsmtFinType2']=data['BsmtFinType2'].fillna(data['BsmtFinType2'].value_counts().index[0])
    data['BsmtFinSF2']=data['BsmtFinSF2'].fillna(data['BsmtFinSF2'].median())
    data['BsmtUnfSF']=data['BsmtUnfSF'].fillna(data['BsmtUnfSF'].median())
    data['TotalBsmtSF']=data['TotalBsmtSF'].fillna(data['TotalBsmtSF'].median())
    data['KitchenQual']=data['KitchenQual'].fillna(data['KitchenQual'].value_counts().index[0])
    data['Functional']=data['Functional'].fillna(data['Functional'].value_counts().index[0])
    data['MSZoning']=data['MSZoning'].fillna(data['MSZoning'].value_counts().index[0])
    data['Utilities']=data['Utilities'].fillna(data['Utilities'].value_counts().index[0])
    data['Exterior1st']=data['Exterior1st'].fillna(data['Exterior1st'].value_counts().index[0])
    data['Exterior2nd']=data['Exterior2nd'].fillna(data['Exterior2nd'].value_counts().index[0]) 
    data['GarageType']=data['GarageType'].fillna(data['GarageType'].value_counts().index[0])
    data['GarageYrBlt']=data['GarageYrBlt'].fillna(data['GarageYrBlt'].median())
    data['GarageFinish']=data['GarageFinish'].fillna(data['GarageFinish'].value_counts().index[0])
    data['GarageCars']=data['GarageCars'].fillna(data['GarageCars'].median())
    data['GarageArea']=data['GarageArea'].fillna(data['GarageArea'].median())
    data['GarageQual']=data['GarageQual'].fillna(data['GarageQual'].value_counts().index[0])
    data['GarageCond']=data['GarageCond'].fillna(data['GarageCond'].value_counts().index[0])
    data['Electrical']=data['Electrical'].fillna(data['Electrical'].value_counts().index[0])
    return data
```

```python
X = train_dataset.iloc[:, :-1].values
y = train_dataset.iloc[:, -1].values
test_dataset = test_dataset.values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
# 1 OneHot / 4-6 OneHot / 7 Label / 8-13 OneHot / 18-22 OneHot / 24-25 Label / 26 OneHot / 27-28 Label / 29 OneHot / 31 OneHot
# 35 OneHot / 36-37 Label / 38 OneHot / 41 Label / 42 OneHot / 44 OneHot / 46 OneHot / 49-51 Label / 
```

```python
le_train = LabelEncoder()
ct_train = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,4,5,6,8,9,10,11,12,13,18,19,20,21,22,26,29,31,35,38,42,44,46])],remainder='passthrough')

le_test = LabelEncoder()
ct_test = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,4,5,6,8,9,10,11,12,13,18,19,20,21,22,26,29,31,35,38,42,44,46])],remainder='passthrough')

le_columns = [7,24,25,27,28,36,37,41,49,50,51]

for i in le_columns:
    X[:, i] = le_train.fit_transform(X[:, i])
X = ct_train.fit_transform(X)

for i in le_columns:
    test_dataset[:, i] = le_test.fit_transform(test_dataset[:, i])
test_dataset = ct_test.fit_transform(test_dataset)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100,criterion='squared_error',random_state=1).fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score

y_pred = model.predict(X_test)
print(model.score(X_test,y_test))
cross_score = cross_val_score(model, X_test, y_test, cv=10)
print(f'Cross val score for our model: {cross_score.mean()}')
```
