```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
test_dataset2 = test_dataset
```

```python
train_dataset
```

```python
test_dataset
```

```python
def clean(data):
    data = data.drop(['Cabin','Ticket', 'Name', 'PassengerId'],axis=1)
    data['Age'] =  data['Age'].fillna(data['Age'].median())
    data['Embarked']=data['Embarked'].fillna("S")     
    data['Fare'] = data['Fare'].fillna(data['Fare'].median())
    return data
```

```python
train_dataset = clean(train_dataset)
train_dataset
```

```python
test_dataset = clean(test_dataset)
test_dataset
```

```python
test = test_dataset.values
```

```python
X = train_dataset.iloc[:, 1:].values
y = train_dataset.iloc[:, 0].values 
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

le_test = LabelEncoder()
le_train = LabelEncoder()
test[:, 1] = le_test.fit_transform(test[:, 1])
X[:, 1] = le_train.fit_transform(X[:, 1])

ct_test = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[-1])], remainder='passthrough')
ct_train = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[-1])], remainder='passthrough')
test = ct_test.fit_transform(test)
X = ct_train.fit_transform(X)
```

```python
test
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=420)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
classifier = SVC(kernel='linear', C=1).fit(X_train, y_train)
y_pred = classifier.predict(X_test)
score = classifier.score(X_test, y_test)
cm = confusion_matrix(y_test, y_pred)
print(f"{cm}, Score {score}")
```

```python
y_prediction = classifier.predict(test)
submission = pd.DataFrame({
    "PassengerId" : test_dataset2["PassengerId"],
    "Survived" : y_prediction
})
submission.head(50)
```
