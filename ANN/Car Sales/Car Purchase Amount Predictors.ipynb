```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import tensorflow.keras as keras
```

```python
dataset = pd.read_csv('Car_Purchasing_Data.csv', encoding='ISO-8859-1')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
ann = keras.models.Sequential()
```

```python
ann.add(keras.layers.Dense(units=128, activation='relu'))
ann.add(keras.layers.Dense(units=64, activation='relu'))
```

```python
ann.add(keras.layers.Dense(units=1, activation='linear'))
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model = ann.fit(X_train, y_train, epochs=300, batch_size=25,verbose=1, validation_split=0.2)
```

```python
plt.plot(model.history['loss'])
plt.show()
```

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```
